## Preprocessing

```python
# Importing dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Importing and reading charity_data.csv
import pandas as pd
application_df = pd.read_csv('https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv')
application_df.head()
```

```python
# Dropping the non-beneficial ID columns 'EIN' and 'NAME'
application_df = application_df.drop(['EIN', 'NAME'], axis = 1)
```

```python
# Determining the number of unique values in each column
unique_values = application_df.nunique()
print(unique_values)
```

```python
# Looking at APPLICATION_TYPE value counts to identify and replace with "Other"
application_type_value_counts = application_df['APPLICATION_TYPE'].value_counts()
print(application_type_value_counts)
```

```python
# Creating a dataframe of the APPLICATION_TYPE value counts
application_type_value_counts_df = application_type_value_counts.reset_index()
application_type_value_counts_df.columns = ['APPLICATION_TYPE', 'count']
application_type_value_counts_df
```

```python
# Choosing a cutoff value and creating a list of application types to be replaced
# using the variable name 'application_types_to_replace'
filtered_application_type_value_counts_df = application_type_value_counts_df[application_type_value_counts_df['count'] < 16]
application_types_to_replace = filtered_application_type_value_counts_df['APPLICATION_TYPE'].tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app, 'Other')

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Looking at CLASSIFICATION value counts to identify and replace with "Other"
classification_value_counts = application_df['CLASSIFICATION'].value_counts()
print(classification_value_counts)
```

```python
# Looking at CLASSIFICATION value counts >1
print(classification_value_counts[classification_value_counts > 1])
```

```python
# Creating a dataframe of the CLASSIFICATION value counts
classification_value_counts_df = classification_value_counts.reset_index()
classification_value_counts_df.columns = ['CLASSIFICATION', 'count']
classification_value_counts_df
```

```python
# Choosing a cutoff value and creating a list of classifications to be replaced
# using the variable name 'classifications_to_replace'
filtered_classification_value_counts_df = classification_value_counts_df[classification_value_counts_df['count'] == 1]
classifications_to_replace = filtered_classification_value_counts_df['CLASSIFICATION'].tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls, 'Other')

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Looking at ASK_AMT value counts
ask_amt_value_counts = application_df['ASK_AMT'].value_counts()
print(ask_amt_value_counts)
```

```python
# Creating categories for ASK_AMT, similar to INCOME_AMT ranges

# Defining a function with multiple conditions
def ask_amt_cat(ask_amt):
    if ask_amt == 5000:
        return '5000'
    elif ask_amt < 15000:
        return '5001-15000'
    elif ask_amt < 50000:
        return '15001-50000'
    elif ask_amt < 150000:
        return '50001-150000'
    elif ask_amt < 450000:
        return '150001-450000'
    elif ask_amt < 1500000:
        return '450001-1.5M'
    elif ask_amt < 10000000:
        return '1.5M-10M'
    elif ask_amt < 100000000:
        return '10M-100M'
    else:
        return '100M+'

# Applying the function to application_df to create a new column 'ASK_AMT_CATS'
application_df['ASK_AMT_CAT'] = application_df['ASK_AMT'].apply(ask_amt_cat)
application_df
```

```python
# Deleting original 'ASK_AMT' column
del application_df['ASK_AMT']
application_df
```

```python
# Converting categorical data to numeric with `pd.get_dummies`
application_df_dummies = pd.get_dummies(application_df)
application_df_dummies
```

```python
# Splitting preprocessed data into features and target arrays
y = application_df_dummies["IS_SUCCESSFUL"].values
X = application_df_dummies.drop(["IS_SUCCESSFUL"], axis = 1).values

# Splitting the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
# Creating a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Checking the number of input features
print(len(X_train[0]))
```

## Optimizing the Model with `keras-tuner`

```python
# Installing keras-tuner
!pip install keras-tuner
```

```python
# From Class 2, Activity 5, "Stu_TuneUp"
# Creating a method that creates a new Sequential model with hyperparameter options
def create_model(hp):
    nn_model = tf.keras.models.Sequential()

    # Allowing kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation', ['relu', 'tanh'])

    # Allowing kerastuner to decide number of neurons in first layer
    nn_model.add(tf.keras.layers.Dense(units = hp.Int('first_units',
        min_value = 5,
        max_value = len(X_train[0]),
        step = 5), activation = activation, input_dim = len(X_train[0])))

    # Allowing kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 5)):
        nn_model.add(tf.keras.layers.Dense(units = hp.Int('units_' + str(i),
            min_value = 5,
            max_value = len(X_train[0]),
            step = 5),
            activation = activation))

    nn_model.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))

    # Compiling the model
    nn_model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

    return nn_model
```

```python
# Importing the kerastuner library
import keras_tuner as kt

tuner = kt.Hyperband(
    create_model,
    objective = 'val_accuracy',
    max_epochs = 20,
    hyperband_iterations = 2)
```

```python
# Running the kerastuner search for best hyperparameters
tuner.search(X_train_scaled, y_train, epochs = 20, validation_data = (X_test_scaled, y_test))
```

```python
# Get top 3 model hyperparameters and print the values
top_hyper = tuner.get_best_hyperparameters(3)
for param in top_hyper:
    print(param.values)
```

```python
# Evaluate the top 3 models against the test dataset
top_model = tuner.get_best_models(3)
for model in top_model:
    model_loss, model_accuracy = model.evaluate(X_test_scaled, y_test, verbose = 2)
    print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Again displaying the top model hyperparameters for compiling, training, and evaluating the model below
top_hyper = tuner.get_best_hyperparameters(1)
for param in top_hyper:
    print(param.values)
```

## Compile, Train and Evaluate the Model

```python
# Defining the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

# Using the hyperparameters from above for this optimization trial (#3), with some minor tweaks
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 65
hidden_nodes_layer2 = 80
hidden_nodes_layer3 = 95

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units = hidden_nodes_layer1, input_dim = number_input_features, activation = 'tanh'))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units = hidden_nodes_layer2, activation = 'tanh'))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units = hidden_nodes_layer3, activation = 'tanh'))

# Output layer
nn.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))

# Checking the structure of the model
nn.summary()
```

```python
# Compiling the model
nn.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
# Assignment prompt: "Create a callback that saves the model's weights every five epochs."

# Querying "How do I create a callback that saves a neural network model's weights every five epochs in Google Colab?"
# in ChatGPT yielded the following/the foundations for the following:

# Step 1: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Step 2: Import necessary libraries and define the custom checkpoint callback
from tensorflow.keras.callbacks import ModelCheckpoint

class CustomModelCheckpoint(ModelCheckpoint):
    def __init__(self, *args, **kwargs):
        self.epochs_since_last_save = 0
        self.period = kwargs.pop('period', 1)
        super().__init__(*args, **kwargs)

    def on_epoch_end(self, epoch, logs = None):
        self.epochs_since_last_save += 1
        if self.epochs_since_last_save >= self.period:
            self.epochs_since_last_save = 0
            super().on_epoch_end(epoch, logs)

# Define the custom checkpoint callback with period=5
checkpoint = CustomModelCheckpoint(
    filepath='/content/drive/My Drive/model_weights_epoch_{epoch:02d}.h5',    # Path in Google Drive
    save_weights_only = True,                                                 # Save only the weights
    period = 5                                                                # Save every 5 epochs
)
```

```python
# Training the model
fit_model = nn.fit(X_train_scaled, y_train, epochs = 50, callbacks = [checkpoint])
```

```python
# Evaluating the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose = 2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
